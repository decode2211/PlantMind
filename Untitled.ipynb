```python
import os

# 1. Create a dictionary mapping the current file names to the new desired file names
rename_mapping = {
    "centrifugal pump": "pump_centrifugal_manual.pdf",
    "Submersible Pump": "pump_submersible_manual.pdf",
    "Induction Motor": "motor_induction_manual.pdf",
    "Air Compressor": "compressor_air_manual.pdf",
    "Refrigerated Air Dryer": "dryer_refrigerated_manual.pdf",
    "Centrifugal Chiller": "chiller_centrifugal_manual.pdf",
    "Cooling Tower": "tower_cooling_manual.pdf",
    "Scroll Chiller": "chiller_scroll_manual.pdf",
    "Power Transformer": "transformer_power_manual.pdf",
    "Main Distribution Panel": "panel_distribution_manual.pdf",
    "Steam Boiler": "boiler_steam_manual.pdf",
    "Belt Conveyor": "conveyor_belt_manual.pdf"
}

# 2. Iterate and rename files in the same directory
success_count = 0

print("Starting file renaming process...\n" + "-"*50)

for old_name, new_name in rename_mapping.items():
    # Append .pdf if the file name keys don't already have it
    old_file_name = old_name if old_name.endswith('.pdf') else f"{old_name}.pdf" 
    
    # Check if the file exists in the current folder before renaming
    if os.path.exists(old_file_name):
        os.rename(old_file_name, new_name)
        print(f"✅ Renamed: '{old_file_name}' ➡️ '{new_name}'")
        success_count += 1
    else:
        print(f"❌ File not found: '{old_file_name}' (Skipped)")

print("-"*50 + f"\nProcess complete. Successfully renamed {success_count} files.")
```
